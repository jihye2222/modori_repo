```python
import pandas as pd
```

```python
ddd = pd.read_csv('C:/Dev/project/influencer_data.csv',encoding='utf-8-sig')
```

```python
ddd
```

```python
from django.db.models import Max
```

```python
influencer_df = ddd.copy()
influencer_df = influencer_df.groupby('ID').agg({'like_cnt': 'sum', 'comments': 'sum', 'followers': 'first'}).reset_index()
influencer_df['points'] = (influencer_df['like_cnt'] + influencer_df['comments']) / influencer_df['followers']
influencer_df = influencer_df.sort_values(by='points', ascending=False)
influencer_df['influencer_score'] = [max(100 - i*2,0) for i in range(len(influencer_df))]
influencer_df.drop('points', axis=1, inplace=True)
influencer_df['influencer_score'] = influencer_df['influencer_score'].apply(lambda x: round(x)).astype(int)
influencer_df = influencer_df[['ID','influencer_score']]


ad_df = ddd.copy()
ad_df = ad_df[ad_df['content'].str.contains('광고|협찬|비즈니스|협업|체험') | ad_df['hashtags'].str.contains('광고|협찬|비즈니스|협업|체험')]
ad_df = ad_df.groupby('ID').agg({'like_cnt': 'sum', 'comments': 'sum', 'followers': 'first'}).reset_index()
ad_df['points'] = (ad_df['like_cnt'] + ad_df['comments']) / ad_df['followers']
ad_df = ad_df.sort_values(by='points', ascending=False)
ad_df['ad_score'] = [max(100 - i *2,0) for i in range(len(ad_df))]
ad_df['ad_score'].fillna(0, inplace=True)  # None 값을 0으로 대체
ad_df = ad_df[["ID","ad_score"]]
```

```python
du = ddd.drop_duplicates(subset=['ID','cate1','cate2','cate3','local'])
```

```python
du = du[['ID','cate1','cate2','cate3','local']]
```

```python
merge_df = pd.merge(influencer_df,ad_df,how='left',on='ID')
```
